```python
import sentencepiece as spm
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Hindi font load
font_path = "NotoSansDevanagari-Regular.ttf"
font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Noto Sans Devanagari'

sp = spm.SentencePieceProcessor()
sp.load("hindi_tokenizer.model")

input_file = "data/hindi_no_exact_dup.txt"
token_freq = Counter()

max_lines = 100_000

with open(input_file, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i == max_lines:
            break
        line = line.strip()
        if line:
            token_ids = sp.encode_as_ids(line)
            token_freq.update(token_ids)

# Top 20
most_common = token_freq.most_common(30)
tokens = [sp.id_to_piece(t[0]) for t in most_common]
counts = [t[1] for t in most_common]

print(tokens, counts)

plt.figure(figsize=(12,6))
plt.bar(tokens, counts)
plt.xticks(rotation=90)
plt.title("Top 20 Most Frequent Tokens (Hindi)")
plt.show()
```

```python
from collections import Counter

sample_texts = [
    "भारत के लोग",
    "Bro mujhe code review kar do please",
    "कंप्यूटरीकरण",
    "yeh text hai Hinglish me",
]

for text in sample_texts:
    tokens = sp.encode(text, out_type=int)
    token_strs = [sp.id_to_piece(t) for t in tokens]
    unk_count = token_strs.count("<unk>")
    print(f"Text: {text}")
    print(f"Tokens: {token_strs}")
    print(f"<unk> count: {unk_count}\n")
```

## exploring the .bin train file data 

```python
# Normal text data:
# आज मौसम कैसा है?
# मैंने फोन खरीदा।
# 😂🚀

# यह UTF-8 string form me stored hota है, CPU ko हर बार:
# ✔️ UTF-8 decode karna hota
# ✔️ Tokenize karna hota
# ✔️ Vocabulary lookup karna hota
# Bahut slow.
```

```python
# .bin file (LLM format)
# यह raw uint16 integers ka continuous stream hota है:
# 234  981  45  1023  3  2201  551  3 ...

# हर number = ek token ID
# 3 = EOS token
# Binary format ke benefits:
# ✔️ Zero decoding
# ✔️ Zero tokenization during training
# ✔️ GPU ko direct memory block milta है
# ✔️ ~10× faster training
# ✔️ Sequential reading perfect for next-token prediction

# 👉 यही OpenAI GPT-2, GPT-3, Meta Llama, Mistral sab use करते हैं.
```
